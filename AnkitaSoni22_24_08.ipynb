```python
N = int(input("Enter the length of list:"))
numbers = input("Enter {} numbers to form list: ".format(N))
test_list = numbers.split()
test_list = [int(i) for i in test_list]
test_list_1 = test_list[::-1]
result = []
for i1, i2 in zip(test_list, test_list_1):
    result.append(i1 * i2)
print("The resulted list: ", result)
```

```python
N= int(input("Enter the number:"))
def factorial(N):

    if N == 0:
        return 1
    result = 1
    for i in range(1, N + 1):
        result *= i
    return result

print(factorial(N))
```
